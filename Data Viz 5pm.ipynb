# Scatter Plot
It can be ussed for outlier detection, understanding the relationship between two continuous variables and understanding the patters etc.

```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
data = px.data.iris()
```

```python
data.head()
```

```python
sns.scatterplot(data['sepal_width'],data['sepal_length'],hue=data['species'])
```

```python
px.scatter(data, x="sepal_width", y="sepal_length", color="species",
           size='petal_length', hover_data=['petal_width'])
```

```python
px.scatter(data, x="sepal_width", y="sepal_length", color="species",
           marginal_y = 'violin',marginal_x="box")
```

```python
data = px.data.tips()
```

```python
data.head()
```

```python
px.scatter(data,
          x="total_bill",
          y = "tip",
          color="size")
```

```python
px.scatter(data,
          x="total_bill",
          y = "tip",
          color="size",
          color_continuous_scale=px.colors.sequential.Viridis)
```

```python
df = px.data.tips()
df.head()
```

```python
df
```

```python
# Facet Grids
fig = px.bar(df, x="sex", y="total_bill", color="smoker", barmode="group",
             facet_row="time", facet_col="day",
             category_orders={"day": ["Thur", "Fri", "Sat", "Sun"],
                              "time": ["Lunch", "Dinner"]})
fig.show()
```

```python
px.scatter(df, x="tip", y="total_bill", color="smoker",
             facet_row="time", facet_col="day",
             category_orders={"day": ["Thur", "Fri", "Sat", "Sun"],
                              "time": ["Lunch", "Dinner"]})
```

```python
data = px.data.wind()
data.head()
```

https://plotly.com/python/polar-chart/

```python
# Scatter Polar Chart
px.scatter_polar(data,r="frequency",theta="direction")
```

```python
px.scatter_polar(data,r="frequency",theta="direction",color="strength")
```

```python
# Line Polar Chart
px.line_polar(data,r="frequency",theta="direction",color="strength")
```

```python
px.line_polar(data, r="frequency", theta="direction", color="strength", line_close=True,
                    color_discrete_sequence=px.colors.sequential.Plasma_r,
                    template="plotly_dark",)
```

```python
# Bar Polar Chart
px.bar_polar(data,r="frequency",theta="direction",color="strength")
```
